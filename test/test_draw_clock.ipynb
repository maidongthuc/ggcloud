```python
from PIL import Image, ImageDraw, ImageFont

# Đường dẫn ảnh gốc
image_path = "/home/acer/Work/ai-research/detection/static/images/raw/checklist/checklist_clock_0_1752133474.png"


try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf", size=30)
except:
    font = ImageFont.load_default()

img = Image.open(image_path)
w,h = img.size[:2]
print(f"Image size: {w}x{h}")
draw = ImageDraw.Draw(img)

points = [
  {"point": [370, 200], "label": "0.4"},
  {"point": [490, 170], "label": "0.5"},
  {"point": [620, 180], "label": "0.6"},
  {"point": [730, 250], "label": "0.7"},
    {"point": [820, 350], "label": "0.8"},
    {"point": [610, 270], "label": "nt"},
        {"point": [495, 530], "label": "nb"},


]
for item in points:
    x, y = item["point"]
    x = int(x / 1000 * w)
    y = int(y / 1000 * h)
    label = item["label"]
    r = 6
    draw.ellipse((x - r, y - r, x + r, y + r), fill="red")
    draw.text((x - 25, y - 40), label, fill="red", font=font)

# Hiển thị trong notebook
img
```
